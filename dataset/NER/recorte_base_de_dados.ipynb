```python
import pandas as pd
pd.set_option('display.max_rows', None)
```

```python
df = pd.read_csv('../data/merged_all_columns_2019_ate_2023-003.csv', encoding='utf-8')
```

# 10 tipos de registro com mais ocorrências

```python
# filtra por relatos de marabá
df_maraba = df[df['municipios'].str.upper() == 'MARABA']

# Agrupa os registros pelo campo 'consolidado' e conta a quantidade de registros em cada grupo
agrupamento = df_maraba.groupby('consolidado').size().reset_index(name='quantidade')

# Ordena o resultado pela maior quantidade (ordem decrescente)
agrupamento = agrupamento.sort_values(by='quantidade', ascending=False)

# Exibe o resultado
print(agrupamento.head(10))
```

```python
# Filtra os registros para o município 'MARABA'
df_maraba = df[df['municipios'].str.upper() == 'MARABA']

# Agrupa os registros pelo campo 'consolidado' e conta a quantidade de ocorrências em cada grupo
agrupamento = df_maraba.groupby('consolidado').size().reset_index(name='quantidade')

# Ordena o agrupamento pela coluna 'quantidade' em ordem decrescente e seleciona os top 10
top_10_consolidados = agrupamento.sort_values(by='quantidade', ascending=False).head(10)

# Obtém a lista dos top 10 valores de 'consolidado'
top_10_list = top_10_consolidados['consolidado'].tolist()

# Filtra o DataFrame original (de Marabá) para obter todas as linhas (todas as colunas) 
# cujo valor em 'consolidado' está entre os top 10
df_top10 = df_maraba[df_maraba['consolidado'].isin(top_10_list)]

print('Quantidade de registros: '+ str(len(df_top10)))

# Gera um novo arquivo CSV com todas as colunas dos registros filtrados
df_top10.to_csv('../data/top_10_consolidado_maraba_full.csv', index=False, encoding='utf-8')

print("Arquivo 'top_10_consolidado_maraba_full.csv' gerado com sucesso!")
```
